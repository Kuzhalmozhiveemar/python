```python
import numpy as np
sns.set(style='white')
rs=np.random.RandomState(10)
mean=[2,2]
cov=[(1,.5),(.5,1)]
x1,x2=rs.multivariate_normal(mean,cov,500).T
x1=pd.Series(x1,name='ExampleX')
x2=pd.Series(x2,name='ExampleY')
ex3=sns.jointplot(x1,x2,kind='kde',space=0)
plt.show()
```
